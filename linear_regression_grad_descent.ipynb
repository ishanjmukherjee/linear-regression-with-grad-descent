# Imports

```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt

print("imported numpy, matplotlib and pyplot")
```

# Create linear dataset with noise


```python
# parameters specifying data
wTrue = 2.
bTrue = 100.
numDatapoints = 50
noiseStdDev = 10

# create dataset
X = np.arange(numDatapoints)
Y = wTrue * X + bTrue + np.random.normal(0, noiseStdDev, numDatapoints)

# plot data in red
plt.plot(X, Y, 'rx')
plt.show()
```

# Batch gradient descent

```python
# hyperparameters
lr = 0.001 # learning rate
epochs = 10000

# training
w, b = 0, 0
for i in range(epochs):
  w, b = (w - lr * 1/numDatapoints * np.sum(((w*X + b) - Y) * X),
          b - lr * 1/numDatapoints * np.sum(((w*X + b) - Y)))

# print weight & bias used to create data, and weight & bias found
print(f"wTrue : {wTrue : 6.2f} | bTrue : {bTrue : 6.2f}")
print(f"    w : {w     : 6.2f} |     b : {b     : 6.2f}")

# plot model in blue and data in red
fig, ax = plt.subplots()
ax.plot(X, w*X + b, 'b', X, Y, 'rx')
plt.show()
```
